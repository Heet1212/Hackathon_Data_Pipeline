```python
import pandas as pd

file_path = '/content/Employee Sample Data.csv'

# Load the CSV file into a DataFrame
df = pd.read_csv(file_path, encoding='latin1')

df.insert(0, 'id', range(1, len(df) + 1))

df
```

```python
#gender insights

gender_groups = df.groupby('Gender')

gender_counts = gender_groups.size()

print(gender_counts)
```

```python
df['Exit Date'].fillna('working', inplace=True)
```

```python
# Check for duplicate rows based on the "EEID" column
duplicates = df[df.duplicated(subset='EEID', keep='first')]

# Remove duplicate rows based on the "EEID" column
df.drop_duplicates(subset='EEID', keep='first', inplace=True)
```
